```python
from tqdm import tqdm
```

```python
dataset = []

for (parent, children, files) in tqdm(os.walk(dataset_path)):
    if len(files) > 0:
        for file_name in files:
            file_path = f"{parent}/{file_name}"
            file_obj = open(file_path, 'r')
            dataset.extend(file_obj.readlines())
```

```python
n_gram = 2
```

```python
dataset = " ".join(dataset)
```

```python
import re, string
pattern = re.compile('[\W_]+')
clean_dataset = pattern.sub(dataset, string.printable)
```
